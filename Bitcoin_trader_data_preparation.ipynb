```python
df1['date'] = pd.to_datetime(df1['date'])

# Extract year and get unique values
unique_years = df1['date'].dt.year.unique()
print(unique_years)
```

```python
#merging both the files for combined analysis
df_merged = pd.merge(df, df1, on='date', how='inner')
```

```python
#standardising the date column format for both the tables
df['date'] = pd.to_datetime(df['date'])
df1['date'] = pd.to_datetime(df1['date'])
```

```python
#checking the first five rows of the merged data
df_merged.head()
```

```python
#getting initial statistical insights
df_merged.describe()
```

```python
#renaming the account numbers to something more simplistic
# Create a mapping from original to simplified names
unique_accounts = df_merged['Account'].unique()
account_map = {acct: f"acct_{i+1}" for i, acct in enumerate(unique_accounts)}

# Apply the mapping
df_merged['account_label'] = df_merged['Account'].map(account_map)
```

```python
#checking whether we got correct account labels by the correct accout
account_info = df_merged[['account_label', 'Account']].drop_duplicates()
print(account_info)
```

```python
#saving this file as a csv
df_merged.to_csv('modified_file.csv', index=False)
```
